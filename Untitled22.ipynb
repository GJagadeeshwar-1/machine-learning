```python
print("This is the program to find even or odd")
number=int(input("Enter a number:"))
if number%2==0:
  print("This number is even")
else:
  print("This number is odd")
print()
print("This is the program to find largest number")
number1=int(input("Enter first number:"))
number2=int(input("Enter seond number:"))
number3=int(input("Enter third number:"))
print("The largest number is ",max(number1,number2,number3))
print()
print("This is the program for the maultiplication table of a number(1-10)")
n=int(input("Enter a table number:"))
print("The multiplication table of a number")
for i in range(1,11):
  print(n,"x",i,"=",n*i)
print()
print("This is the program to finding maximum,minimum,sum of elements and removing duplicates")
number=[1,2,2,3,4,4,5]
number1=list(set(number))
print("The maximum number is ",max(number1))
print("The minimum number is ",min(number1))
print("The sum of elements is ",sum(number1))
print("Before the removing duplicate elements",number)
print("After the removing duplicate elements",number1)
print()
print("This is the program to find the highest marks in dictionary with the students marks")
students_marks={'ram':60,'sham':50,'ravi':70,'harish':90}
print("The highest marks is ",max(students_marks.values()))
print()
```
